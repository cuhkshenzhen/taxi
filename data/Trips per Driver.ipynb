```python
for i in range(3, 31):
    if i not in [7, 13, 17]:
        filename = filename_syntax.format(i)
        print("{0/6} Process:", filename)
        print("(1/6) Reading file", "at", datetime.now())
        df = pd.read_csv(filename, header=None, names=["TaxiID", "Latitude", "Longtitude", "Carriage", "Time"], parse_dates=[4], infer_datetime_format=True)
        
        print("{2/6} Readed. Sorting", "at", datetime.now())
        df.sort_values("Time", inplace=True)
        
        print("(3/6) Sorted. Grouping", "at", datetime.now())
        g = df.groupby("TaxiID")
        
        print("(4/6) Grouped. Getting description", "at", datetime.now())
        desc = g.describe()
        
        print("(5/6) Description got. Saving", "at", datetime.now())
        desc.to_html("Summary/html/" + str(i) + ".html")
        desc.to_json("Summary/json/" + str(i) + ".json")
        desc.to_pickle("Summary/pickle/" + str(i) + ".pickle")
        del df
        del g
        del desc
        
        print("(6/6) Finished.", "at", datetime.now())
        
        
        
```
